```python

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
 
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


faces = faceCascade.detectMultiScale(gray, 1.1,4)


for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), ( 255, 0, 0), 2)
```

```python
plot.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX

# Define the parameters as a dictionary
text_params = {
    'img': img,                                   # Image on which text will be inserted
    'text': predictions[0]['dominant_emotion'],   # Text to be inserted (dominant emotion)
    'org': (1200, 1100),                              # Position where text will be placed (x, y)
    'fontFace': font,                             # Font type
    'fontScale': 8,                               # Font scale
    'color': (0, 0 , 255),                         # Color of the text (BGR format)
    'thickness': 6,                               # Thickness of the text
    'lineType': cv2.LINE_4                        # Line type
}

# Use putText() method for inserting text on the image
cv2.putText(**text_params)
```

```python
plot.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plot.axis('off')
plot.show()
```
